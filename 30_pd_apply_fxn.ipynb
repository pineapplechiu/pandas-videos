```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

## map

```python
# dummy variable for sex: translate male and female to 1 and 0
# map existing value of a series to a different set of values
# map to dictionary
train['Sex_num'] = train.Sex.map({'female': 0, 'male': 1})
```

```python
# see rows 0 to 4, columns sex and sex_num
train.loc[0:4, ['Sex','Sex_num']]
```

## apply as a series method

```python
# apply is both a series method and dataframe method
```

```python
# calculate length of each string in column, create new column with integer value of length
# as a series method

# new column
train['Name_length'] = train.Name.apply(len)
# only pass name of function, not len with parenthesis, e.g. len()
```

```python
# review first 5 rows again
# for columns name and name_length
train.loc[0:4, ['Name','Name_length']]
```

```python
import numpy as np
```

```python
# round fare column up 
# create new column called fare_ceil
train['Fare_ceil'] = train.Fare.apply(np.ceil)
# apply ceil function from numpy
```

```python
# review first 5 rows again
# for columns fare and fare_ceil
train.loc[0:4, ['Fare','Fare_ceil']]
```

```python
# extract last name of each column to its own column
# use string method to split name on commas
train.Name.str.split(',').head()
```

```python
# now split is a list of strings, meaning
# series of python lists
# each list is made up of strings
# first element comma second element
```

```python
# really only want the first list element from each series
# write a function 
def get_element(my_list, position):
    return my_list[position]
```

```python
# apply this function, keyword argument is position = 0
train.Name.str.split(',').apply(get_element, position=0)
```

```python
train.Name.str.split(',').apply(get_element, position=0).head()
```

```python
# use lambda function
train.Name.str.split(',').apply(lambda x: x[0]).head()
```

## apply as a dataframe method

```python
# use another data set for this demo
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
# applies function along either axis of the dataframe
# use subset of dataframe with all rows and only beer through wine servings
drinks.loc[:, 'beer_servings':'wine_servings']
```

```python
# apply to entire dataframe
# axis 0 is column direction
# apply python's max function
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=0)
# figure out max value of each of those columns
```

```python
# apply to entire dataframe
# axis 1 is each row direction
# apply python's max function
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=1)
# max value of those three columns
```

```python
# which column is the the max?
drinks.loc[:, 'beer_servings':'wine_servings'].apply(np.argmax, axis=1)
```

## applymap

```python
# applymap is also a dataframe method
# applies to every element of the dataframe
drinks.loc[:, 'beer_servings':'wine_servings'].applymap(float)
# changes every element to a floating point 
```

```python
# can be used to overwrite each existing column in dataframe
drinks.loc[:, 'beer_servings':'wine_servings'] = drinks.loc[:, 'beer_servings':'wine_servings'].applymap(float)
drinks.head()
# from integers to floating point numbers
```
